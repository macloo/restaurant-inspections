```python
conn = sqlite3.connect("rinspect.sqlite")
df = pd.read_sql_query("select * from fdinsp;", conn)
```

```python
df.sort_values(by='highvio', ascending=False)
```

```python
df['inspdate'] = pd.to_datetime(df['inspdate']) # making sure 'inspdate' is dtype datetime64[ns]
start_date = pd.to_datetime('today') - pd.DateOffset(days=7) #'2018-7-25' # how to select: coded or user input?
end_date = pd.to_datetime('today') # how to select select: today or user input
mask = (df['inspdate'] > start_date) & (df['inspdate'] <= end_date)
df = df.loc[mask]
df.sort_values(by='inspdate', ascending=False)

# showing an emnpy set?
```
